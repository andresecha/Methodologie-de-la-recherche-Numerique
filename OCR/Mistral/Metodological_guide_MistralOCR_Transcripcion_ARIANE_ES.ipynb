### Planteamiento del problema

Las ciencias humanas y sociales trabajan cotidianamente con **documentos que no son texto puro**: artículos escaneados, páginas de archivos fotográficos, manuscritos, actas administrativas, planos con anotaciones, recortes de prensa, partituras, formularios históricos, presentaciones de diapositivas y toda clase de PDF "imagen" en los que el texto no es seleccionable ni buscable.

Mientras estos documentos permanezcan como imágenes opacas, quedan **fuera del alcance de las herramientas de análisis textual, de indexación y de búsqueda** que son la base de buena parte de la investigación contemporánea en humanidades digitales. El reconocimiento óptico de caracteres (OCR) clásico resuelve parte del problema, pero a menudo **pierde la estructura del documento** (encabezados, tablas, ecuaciones, imágenes incrustadas) y produce resultados mediocres en documentos complejos o multilingües.

Surge entonces una **pregunta operativa**:

> *¿Es posible, con las herramientas actuales de OCR basadas en inteligencia artificial, extraer de manera relativamente fiable no solo el texto sino también la estructura, las tablas, las ecuaciones y las imágenes de documentos complejos —impresos, manuscritos o digitalizados— en múltiples lenguas y escrituras, y cómo integrar estos resultados en nuestros flujos de trabajo de investigación?*

### Qué propone este cuaderno

Este cuaderno de Google Colab ofrece una **guía práctica y razonada** para utilizar [**Mistral OCR**](https://mistral.ai/news/mistral-ocr), la API de reconocimiento óptico de documentos desarrollada por [Mistral AI](https://mistral.ai) (empresa francesa). A diferencia de los OCR tradicionales, Mistral OCR es un **modelo de comprensión documental** que no se limita a leer texto: entiende la jerarquía, el diseño y los elementos multimodales de cada página.

| Capacidad | Descripción | Utilidad para la investigación |
|-----------|-------------|-------------------------------|
| **Extracción de texto estructurado** | Produce texto en Markdown conservando encabezados, listas, párrafos y enlaces | Obtener versiones textuales fieles a la organización original del documento |
| **Reconocimiento de tablas** | Extrae tablas en formato Markdown o HTML | Recuperar datos tabulares de informes, censos, inventarios o catálogos |
| **Extracción de ecuaciones** | Reconoce expresiones matemáticas y fórmulas LaTeX | Digitalizar artículos científicos y técnicos conservando las fórmulas |
| **Extracción de imágenes incrustadas** | Identifica y extrae las imágenes contenidas en el documento | Preservar figuras, gráficos y fotografías junto con el texto que las describe |
| **Comprensión multilingüe** | Reconoce miles de lenguas y sistemas de escritura | Procesar fondos documentales en árabe, chino, griego, hebreo, devanagari, etc. |
| **Salida estructurada (JSON)** | Permite definir esquemas de datos (Pydantic) para extraer campos específicos | Automatizar la extracción de metadatos (autor, título, fecha, resumen) de lotes de documentos |
| **Anotaciones** | Genera descripciones y clasificaciones de figuras y del documento completo | Enriquecer automáticamente los metadatos de un corpus documental |

## 📐 Para saber más — Parámetros principales de la API Mistral OCR

> Esta celda es **solo informativa**, no hace falta ejecutarla. Os explica las opciones que podréis ajustar en los pasos siguientes.

La llamada principal se hace mediante `client.ocr.process(...)`. Los parámetros más importantes son:

| Parámetro | Descripción | Ejemplo |
|-----------|-------------|--------|
| `model` | Modelo OCR a utilizar | `'mistral-ocr-latest'` |
| `document` | Describe lo que enviamos al OCR (URL o archivo) | `{'type': 'document_url', 'document_url': '...'}` |
| `include_image_base64` | `True` = recuperar también las imágenes / `False` = solo texto | `True` |
| `pages` | Procesar solo ciertas páginas (los índices empiezan en 0) | `[0, 1, 2]` para pág. 1-3 |
| `image_limit` | Número máximo de imágenes por página | `5` |
| `image_min_size` | Tamaño mínimo de imagen a recuperar (en píxeles) | `100` |

**Estructura de la respuesta (`ocr_response`):**

```
ocr_response
  └── .pages  →  lista de páginas
        ├── .markdown  →  texto estructurado de la página
        └── .images    →  lista de imágenes de la página
              ├── .image_base64  →  imagen codificada en base64
              └── .id            →  identificador de la imagen
```

# 🌐 MODO A — OCR de un PDF disponible en línea (mediante URL)

Usad este modo si vuestro documento es accesible mediante un enlace web (p. ej.: un artículo en acceso abierto, un informe público, un PDF alojado en un servidor).

> 📌 Si vuestro archivo está en el ordenador, pasad directamente al **MODO B** más abajo.

## 🔗 PASO A1 — Indicar la URL del documento

Reemplazad la URL de ejemplo por la de vuestro propio documento PDF. La URL debe empezar por `https://` y apuntar directamente a un archivo `.pdf`.

> 💡 También funciona con imágenes directas (`.png`, `.jpg`) cambiando `type` a `'image_url'` en la celda siguiente.

````python
# ─────────────────────────────────────────────────────────────────────────────
# ✏️  ZONA A MODIFICAR: reemplazad la URL siguiente por la de vuestro documento
# ─────────────────────────────────────────────────────────────────────────────

PDF_URL = 'https://arxiv.org/pdf/2201.04234'  # ← URL de vuestro PDF

# Nombre del archivo de salida (sin extensión) — modificadlo si queréis
NOMBRE_SALIDA = 'resultado_ocr_url'

print(f'📄 Documento objetivo: {PDF_URL}')
print(f'📝 Los resultados se guardarán como: {NOMBRE_SALIDA}.*')
````

## 🚀 PASO A2 — Lanzamiento del OCR

Esta celda envía el documento a Mistral OCR y recupera el resultado. El procesamiento puede tardar varios segundos o un minuto, según el tamaño del PDF.

**Qué hace el código línea por línea:**

- `model='mistral-ocr-latest'` → usa el último modelo OCR de Mistral
- `type='document_url'` → indica que enviamos un enlace (no un archivo local)
- `document_url=PDF_URL` → la URL que definisteis en el paso anterior
- `include_image_base64=True` → solicita recuperar también las imágenes del documento (si no las necesitáis, poned `False` para ir más rápido)

````python
# Envío del documento a la API Mistral OCR
# No modifiquéis este bloque — funciona tal cual

print('⏳ OCR en curso, esperad...')

ocr_response = client.ocr.process(
    model='mistral-ocr-latest',
    document={
        'type': 'document_url',       # tipo de fuente: enlace web
        'document_url': PDF_URL        # la URL del PDF a procesar
    },
    include_image_base64=True          # True = recuperar imágenes / False = solo texto
)

nb_paginas = len(ocr_response.pages)
print(f'✅ OCR terminado! Número de páginas procesadas: {nb_paginas}')
````

## 👁️ PASO A3 — Vista previa del resultado

Mostramos las **primeras páginas** para verificar que el OCR ha funcionado correctamente. El texto se renderiza en Markdown: los títulos, listas y tablas del documento original se conservan de forma estructurada.

> Modificad `NB_PAG_VISTA_PREVIA` si queréis ver más o menos páginas.

````python
# ─────────────────────────────────────────────────────────────────────────────
# ✏️  ZONA A MODIFICAR: número de páginas a mostrar en la vista previa (por defecto: 3)
# ─────────────────────────────────────────────────────────────────────────────
NB_PAG_VISTA_PREVIA = 3

# Bucle sobre las primeras páginas del resultado
for i, pagina in enumerate(ocr_response.pages[:NB_PAG_VISTA_PREVIA]):
    print(f'\n{'─'*50}')
    print(f'  PÁGINA {i + 1}')
    print(f'{'─'*50}')
    display(Markdown(pagina.markdown))  # renderiza el texto Markdown en el notebook
````

## 💾 PASO A4 — Guardado de los resultados

Guardamos el texto extraído en **3 formatos** para adaptarnos a distintos usos:

| Formato | Descripción | Cuándo usarlo |
|---------|-------------|---------------|
| **`.txt`** | Texto plano, todas las páginas concatenadas | Lectura rápida, copiar/pegar |
| **`.md`** | Markdown con formato conservado (títulos, tablas…) | Edición, publicación web |
| **`.json`** | Estructurado, una entrada por página | Tratamiento automático, bases de datos |

Los archivos se guardan en el directorio de trabajo actual del notebook.

````python
# ── Guardado en .txt (texto plano) ──────────────────────────────────────────
# Concatenamos el Markdown de todas las páginas separadas por línea en blanco
texto_completo = '\n\n'.join(pag.markdown for pag in ocr_response.pages)
with open(f'{NOMBRE_SALIDA}.txt', 'w', encoding='utf-8') as f:
    f.write(texto_completo)
print(f'✅ Texto guardado: {NOMBRE_SALIDA}.txt')

# ── Guardado en .md (Markdown con comentarios de página) ─────────────────────
with open(f'{NOMBRE_SALIDA}.md', 'w', encoding='utf-8') as f:
    for i, pag in enumerate(ocr_response.pages):
        f.write(f'<!-- PÁGINA {i+1} -->\n')  # marcador de página
        f.write(pag.markdown + '\n\n')
print(f'✅ Markdown guardado: {NOMBRE_SALIDA}.md')

# ── Guardado en .json (estructurado, una entrada por página) ─────────────────
datos_json = [
    {
        'pagina': i + 1,
        'texto': pag.markdown,
        'nb_imagenes': len(pag.images) if pag.images else 0
    }
    for i, pag in enumerate(ocr_response.pages)
]
with open(f'{NOMBRE_SALIDA}.json', 'w', encoding='utf-8') as f:
    json.dump(datos_json, f, ensure_ascii=False, indent=2)
print(f'✅ JSON guardado: {NOMBRE_SALIDA}.json')

print(f'\n📦 Total: 3 archivos generados con el prefijo "{NOMBRE_SALIDA}"')
````

## 🖼️ PASO A5 (opcional) — Extracción de las imágenes del documento

Si el PDF contiene imágenes (ilustraciones, figuras, tablas escaneadas…), esta celda las decodifica desde el formato `base64` y las guarda como archivos `.png` en una carpeta dedicada.

**¿Qué es base64?** Es una forma de codificar datos binarios (como una imagen) en texto puro para que puedan circular por la red. Aquí simplemente lo decodificamos para recuperar la imagen original.

> **Consejo:** si vuestro documento no contiene imágenes o no las necesitáis, podéis saltaros esta celda.

````python
# Creación de la carpeta de destino para las imágenes
carpeta_imagenes = Path(f'imagenes_{NOMBRE_SALIDA}')
carpeta_imagenes.mkdir(exist_ok=True)  # exist_ok=True: no da error si ya existe

contador_imagenes = 0

# Recorremos cada página del resultado OCR
for i, pag in enumerate(ocr_response.pages):
    if not pag.images:
        continue  # no hay imágenes en esta página → pasamos a la siguiente

    for img in pag.images:
        if not img.image_base64:
            continue  # imagen sin datos base64 → la ignoramos

        # Decodificación: de texto base64 → bytes binarios (la imagen real)
        datos_imagen = base64.b64decode(img.image_base64)

        # Nombre del archivo: pag01_img01.png, pag01_img02.png, etc.
        nombre_img = f'pag_{i+1:02d}_img{contador_imagenes+1:02d}.png'
        ruta_img = carpeta_imagenes / nombre_img

        # Escritura de la imagen en disco
        with open(ruta_img, 'wb') as f:
            f.write(datos_imagen)

        contador_imagenes += 1
        print(f'  🖼️  Imagen guardada: {ruta_img}')

# Resumen final
if contador_imagenes == 0:
    print('ℹ️  No se han encontrado imágenes en este documento.')
else:
    print(f'\n✅ {contador_imagenes} imagen(es) guardada(s) en la carpeta: {carpeta_imagenes}/')
````

# 📁 MODO B — OCR de un archivo PDF local (desde vuestro ordenador)

Usad este modo si el documento está en vuestro ordenador (o en vuestro Google Drive). El proceso incluye un paso adicional respecto al Modo A: **primero subimos el archivo a los servidores de Mistral**, y luego lanzamos el OCR.

El flujo completo del Modo B es el siguiente:

```
Vuestro PDF  →  [B1] Seleccionar archivo
             →  [B2] Subir a Mistral  →  obtener FILE_ID
             →  [B3] Lanzar OCR con FILE_ID
             →  [B4] Vista previa
             →  [B5] Guardar resultados
             →  [B6] Extraer imágenes (opcional)
             →  [B7] Eliminar archivo de los servidores Mistral
```

> ⚠️ El archivo enviado a Mistral se almacenará temporalmente en sus servidores. No transmitáis documentos confidenciales sin revisar las [condiciones de uso de Mistral AI](https://mistral.ai/terms).

## 📤 PASO B1 — Seleccionar vuestro archivo PDF

Dos sub-opciones según vuestro entorno de trabajo:

**En Google Colab:** la celda abre un cuadro de diálogo para cargar el archivo directamente desde vuestro ordenador. Haced clic en el botón que aparece tras la ejecución.

**En Jupyter local:** indicad manualmente la ruta completa hacia vuestro archivo en la variable `RUTA_PDF_LOCAL`.

> 📌 Ejemplos de ruta:
> - Mac/Linux: `'/Users/nombre/Documentos/articulo.pdf'`
> - Windows: `'C:/Users/nombre/Documentos/articulo.pdf'`

````python
# ─────────────────────────────────────────────────────────────────────────────
# ✏️  ZONA A MODIFICAR según vuestro entorno
# ─────────────────────────────────────────────────────────────────────────────

# Sub-opción B1a — En Google Colab: carga interactiva desde el ordenador
try:
    from google.colab import files
    print('📂 Colab detectado. Haced clic en el botón para cargar vuestro PDF...')
    archivos_cargados = files.upload()  # abre el cuadro de diálogo
    RUTA_PDF_LOCAL = list(archivos_cargados.keys())[0]  # nombre del archivo cargado
    print(f'✅ Archivo recibido: {RUTA_PDF_LOCAL}')

# Sub-opción B1b — En Jupyter local: ruta manual
except Exception:
    RUTA_PDF_LOCAL = 'mi_documento.pdf'  # ← reemplazad por vuestra ruta completa
    print(f'📄 Archivo local seleccionado: {RUTA_PDF_LOCAL}')

# Nombre de salida generado automáticamente desde el nombre del archivo (sin extensión)
NOMBRE_SALIDA_LOCAL = Path(RUTA_PDF_LOCAL).stem
print(f'📝 Los resultados se guardarán con el nombre: "{NOMBRE_SALIDA_LOCAL}.*"')
````

## ☁️ PASO B2 — Envío del archivo a los servidores Mistral

Antes de lanzar el OCR, debemos **subir el archivo** a la API Mistral. Funciona como un buzón temporal: Mistral guarda el archivo, nos da un identificador único (`FILE_ID`) y nosotros lo usamos en el paso siguiente para solicitar el OCR.

Este proceso puede tardar algunos segundos según el tamaño del archivo y vuestra conexión.

````python
# Apertura del archivo en modo binario ('rb' = read binary) y envío a Mistral
print(f'⏳ Enviando "{RUTA_PDF_LOCAL}" a Mistral...')

with open(RUTA_PDF_LOCAL, 'rb') as f:
    archivo_subido = client.files.upload(
        file={
            'file_name': Path(RUTA_PDF_LOCAL).name,  # nombre visible del archivo
            'content': f                              # contenido binario del PDF
        },
        purpose='ocr'  # indica a Mistral que este archivo es para OCR (no para fine-tuning)
    )

# Guardamos el identificador único — lo necesitaremos en los pasos siguientes
FILE_ID = archivo_subido.id

print(f'✅ Archivo enviado con éxito!')
print(f'🆔 Identificador Mistral del archivo (FILE_ID): {FILE_ID}')
````

## 🚀 PASO B3 — Lanzamiento del OCR sobre el archivo local

El mecanismo es idéntico al Modo A. La diferencia es que en lugar de una URL pública, usamos una **URL firmada** (de acceso temporal y seguro) generada automáticamente a partir del `FILE_ID` del archivo que acabamos de subir.

**Opciones útiles que podéis activar descomentando las líneas correspondientes:**

- `pages=[0, 1, 2]` → procesar únicamente las páginas 1, 2 y 3 *(índices empiezan en 0)*
- `include_image_base64=False` → no recuperar imágenes (más rápido y económico)
- `image_limit=5` → limitar a 5 imágenes por página como máximo
- `image_min_size=100` → ignorar imágenes de menos de 100 píxeles

````python
# Paso 1: Recuperación de la URL firmada para acceder al archivo subido
# (la URL firmada tiene validez temporal, no hace falta guardarla)
url_firmada = client.files.get_signed_url(file_id=FILE_ID)

# Paso 2: Lanzamiento del OCR
print('⏳ OCR en curso, esperad...')

ocr_response_local = client.ocr.process(
    model='mistral-ocr-latest',
    document={
        'type': 'document_url',
        'document_url': url_firmada.url   # URL temporal hacia el archivo subido
    },
    include_image_base64=True             # True = recuperar imágenes / False = solo texto
    # pages=[0, 1, 2],                    # ← descomentad para procesar solo ciertas páginas
    # image_limit=10,                      # ← descomentad para limitar el nº de imágenes
)

nb_pag_local = len(ocr_response_local.pages)
print(f'✅ OCR terminado! Número de páginas procesadas: {nb_pag_local}')
````

## 👁️ PASO B4 — Vista previa del resultado

Verificación rápida: mostramos las primeras páginas para asegurarnos de que el OCR ha funcionado correctamente antes de guardar.

````python
# ─────────────────────────────────────────────────────────────────────────────
# ✏️  ZONA A MODIFICAR: número de páginas a mostrar en la vista previa
# ─────────────────────────────────────────────────────────────────────────────
NB_PAG_VISTA_PREVIA_LOCAL = 3

for i, pag in enumerate(ocr_response_local.pages[:NB_PAG_VISTA_PREVIA_LOCAL]):
    print(f'\n{'─'*50}')
    print(f'  PÁGINA {i + 1}')
    print(f'{'─'*50}')
    display(Markdown(pag.markdown))
````

## 💾 PASO B5 — Guardado de los resultados

Idéntico al Modo A: guardamos en `.txt`, `.md` y `.json`.

> Los archivos llevan automáticamente el nombre de vuestro PDF original (sin la extensión `.pdf`).

````python
# ── Guardado en .txt (texto plano) ──────────────────────────────────────────
texto_completo_local = '\n\n'.join(pag.markdown for pag in ocr_response_local.pages)
with open(f'{NOMBRE_SALIDA_LOCAL}.txt', 'w', encoding='utf-8') as f:
    f.write(texto_completo_local)
print(f'✅ Texto guardado: {NOMBRE_SALIDA_LOCAL}.txt')

# ── Guardado en .md (Markdown con marcadores de página) ──────────────────────
with open(f'{NOMBRE_SALIDA_LOCAL}.md', 'w', encoding='utf-8') as f:
    for i, pag in enumerate(ocr_response_local.pages):
        f.write(f'<!-- PÁGINA {i+1} -->\n')
        f.write(pag.markdown + '\n\n')
print(f'✅ Markdown guardado: {NOMBRE_SALIDA_LOCAL}.md')

# ── Guardado en .json (estructurado, una entrada por página) ─────────────────
datos_json_local = [
    {
        'pagina': i + 1,
        'texto': pag.markdown,
        'nb_imagenes': len(pag.images) if pag.images else 0
    }
    for i, pag in enumerate(ocr_response_local.pages)
]
with open(f'{NOMBRE_SALIDA_LOCAL}.json', 'w', encoding='utf-8') as f:
    json.dump(datos_json_local, f, ensure_ascii=False, indent=2)
print(f'✅ JSON guardado: {NOMBRE_SALIDA_LOCAL}.json')

print(f'\n📦 Total: 3 archivos generados con el prefijo "{NOMBRE_SALIDA_LOCAL}"')
````

## 🖼️ PASO B6 (opcional) — Extracción de imágenes del documento local

Mismo principio que en el Modo A: recorremos las páginas, decodificamos las imágenes en base64 y las guardamos como archivos `.png` en una carpeta dedicada.

````python
carpeta_imagenes_local = Path(f'imagenes_{NOMBRE_SALIDA_LOCAL}')
carpeta_imagenes_local.mkdir(exist_ok=True)

contador_img_local = 0

for i, pag in enumerate(ocr_response_local.pages):
    if not pag.images:
        continue  # no hay imágenes en esta página

    for img in pag.images:
        if not img.image_base64:
            continue  # imagen sin datos, la ignoramos

        # Decodificación base64 → bytes de imagen
        datos_imagen = base64.b64decode(img.image_base64)

        nombre_img = f'pag_{i+1:02d}_img{contador_img_local+1:02d}.png'
        ruta_img = carpeta_imagenes_local / nombre_img

        with open(ruta_img, 'wb') as f:
            f.write(datos_imagen)

        contador_img_local += 1
        print(f'  🖼️  Imagen guardada: {ruta_img}')

if contador_img_local == 0:
    print('ℹ️  No se han encontrado imágenes en este documento.')
else:
    print(f'\n✅ {contador_img_local} imagen(es) guardada(s) en: {carpeta_imagenes_local}/')
````
